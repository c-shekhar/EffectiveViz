```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '../data/'
```

```python
pokemon_data = pd.read_csv(path+'pokemon/Pokemon.csv',index_col=0)
```

```python
pokemon_data.head()
```

```python
sns.countplot(pokemon_data['Generation'])
```

```python
sns.distplot(pokemon_data['HP'])
```

```python
sns.jointplot(x='Attack',y='Defense',data=pokemon_data)
```

```python
sns.jointplot(x='Attack',y='Defense',data=pokemon_data,kind='hex')
```

```python
sns.kdeplot(pokemon_data['HP'],pokemon_data['Attack'])
```

```python
df = pokemon_data[['Legendary','Attack']]
```

```python
sns.boxplot(x='Legendary',
           y='Attack',
           data=df)
```

```python
sns.violinplot(x='Legendary',
           y='Attack',
           data=df)
```

```python
g = sns.FacetGrid(pokemon_data,row='Legendary')
g.map(sns.kdeplot,'Attack')
```

```python
g = sns.FacetGrid(pokemon_data,row='Generation',col='Legendary')
g.map(sns.kdeplot,'Attack')
```

```python
df = pokemon_data[['HP','Attack','Defense']]
sns.pairplot(df)
```

```python
sns.jointplot(x='Attack',y='Defense',data=pokemon_data)
```

```python
sns.lmplot(x='Attack',y='Defense',data=pokemon_data,fit_reg=False)
```

```python
sns.lmplot(x='Attack',y='Defense',data=pokemon_data,markers=['x','o'],hue='Legendary',fit_reg=False)
```

```python
sns.boxplot(x='Generation',y='Total',data=pokemon_data)
```

```python
sns.boxplot(x='Generation',y='Total',hue='Legendary',data=pokemon_data)
```
