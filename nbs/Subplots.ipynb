```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = '../data/'
```

```python
reviews = pd.read_csv(path+"wine-reviews/winemag-data_first150k.csv", index_col=0)
reviews.head()
```

```python
fig, axarr = plt.subplots(2,1,figsize=(12,8))
```

```python
axarr
```

```python
fig, axarr = plt.subplots(2,1,figsize=(12,8))
reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0])
reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1])
```

```python
fig, axarr = plt.subplots(2,2,figsize=(12,8))
reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0][0])
reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1][1])
```

```python
fig, axarr = plt.subplots(2,2,figsize=(12,8))
reviews['points'].value_counts().sort_index().plot.bar(
    ax=axarr[0][0],fontsize=10,color='mediumvioletred')
axarr[0][0].set_title("Wine Points")
reviews['price'].value_counts().plot.hist(
    ax=axarr[0][1],fontsize=10,color='mediumvioletred')
axarr[0][1].set_title("Wine Prices")
reviews['variety'].value_counts().head(20).plot.bar(
    ax=axarr[1][0],fontsize=10,color='mediumvioletred')
axarr[1][0].set_title("Wine Varieties")
reviews['province'].value_counts().head(20).plot.bar(
    ax=axarr[1][1],fontsize=10,color='mediumvioletred')
axarr[1][1].set_title("Wine Origins")
```
